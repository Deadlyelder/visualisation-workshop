# Uniform Manifold Approximation and Projection (UMAP)

* UMAP is significantly faster than t-SNE
* Generalized embedding dimension: Unlike t-SNE, which is used for visualization, UMAP has no restrictions on the size of embedding dimension, so it can be applied as a general dimensionality reduction algorithm.
* Global structure: better preserves the overall manifold structure
* Solid theoretical background: based on Riemann geometry and topology (according to the author, not knowing the theoretical details is good for mental health)
* Paper: https://arxiv.org/abs/1802.03426

## Dataset

To better understand the how UMAP works, we make use of the MNIST dataset.
For quick reference we download it from the web, otherwise if running locally the dataset is available in the `/data` folder at the root of the repository.

```python
mnist = pd.read_csv('https://raw.githubusercontent.com/sbussmann/kaggle-mnist/master/Data/train.csv')
```

```python
mnist.head()
```

```python
# We remove the label column from our dataset
label = mnist['label']
mnist.drop(['label'], inplace=True, axis=1)
mnist.head()
```

## UMAP result
Lets beign with importing the package

```python
!pip install umap-learn
```

```python
import umap

# We fix our seeder
reducer = umap.UMAP(random_state=42)
mnist_umap = reducer.fit_transform(mnist, label)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

for idx in range(10):
    fig.add_trace(
        go.Scatter(x=mnist_umap[:, 0][label == idx],
                   y=mnist_umap[:, 1][label == idx],
                   name=str(idx),
                   opacity=0.6,
                   mode='markers',
                   marker=dict(color=color[idx])))

fig.update_layout(width=800,
                  height=800,
                  title="UMAP result",
                  yaxis=dict(scaleanchor="x", scaleratio=1),
                  legend=dict(orientation="h",
                              yanchor="bottom",
                              y=1.02,
                              xanchor="right",
                              x=1))

fig.show()
```
